Figure 1. The conflict between spike-timing-dependent plasticity (STDP) and spike synchronization in Hebbian cell assemblies. (A) In standard spike pair-based (doublet) STDP models, the weight change Δw is a function F(Δt) of the time lag Δt: = tpost − tpre between presynaptic and postsynaptic spike. For positive time lags Δt > 0 (pre-before-post) STDP models predict long-term potentiation (LTP) with Δw > 0. For negative time lags Δt < 0 (post-before-pre) STDP models predict long-term depression (LTD) with Δw < 0.
